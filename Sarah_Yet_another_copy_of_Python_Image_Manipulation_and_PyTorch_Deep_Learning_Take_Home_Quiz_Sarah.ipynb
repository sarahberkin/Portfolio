## Create Dataset

```
import torch
from torch.utils.data import Dataset, DataLoader
import cv2
import numpy as np
from torchvision import transforms
```

```
class CardiacMRIDataset(Dataset):
  def __init__(self, image_dir, transform=None):
    self.image_dir = image_dir
    self.image_files = [f for f in os.listdir(image_dir) if f.endswith('.jpg')]
    self.transform = transform

  def __len__(self):
    return len(self.image_files)

  def __getitem__(self, idx):
    img_name = self.image_files[idx]
    image_path = os.path.join(self.image_dir, img_name)
    image = cv2.imread(image_path)

    mask = self.create_binary_mask(image)

    #Convert to RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    #Apply transformation (resize, normalization)
    if self.transform:
      image = self.transform(image)
      mask = torch.tensor(mask, dtype=torch.float32).unsqueeze(0) / 255.0
      resize_transform = None
      for t in self.transform.transforms:
        if isinstance(t, transforms.Resize):
          resize_transform = t
          break
      if resize_transform:
        mask = resize_transform(mask)

    return image, mask

  def create_binary_mask(self, image):
    #Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    #Threshold the image to create a binary mask
    _, thresh = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)
    #Find contours of the regions of interest (left ventricle, right ventricle)
    cnts, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    #Create blank mask
    mask = np.zeros_like(gray)
    #For each contour, draw it on the mask
    for c in cnts:
      cv2.drawContours(mask, [c], -1, 255, thickness=cv2.FILLED)

    return mask
```

## Dataloader

```
#Defining transformations
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((224, 224)),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

#Create dataset instance for images
image_dir = uploaded_files_dir
dataset = CardiacMRIDataset(image_dir, transform=transform)

#Checking dataset size
print(f"Total dataset size: {len(dataset)}")
```

## Start training

```
#Split into training & testing
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

#Create dataloader instances for batch processing
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False)

#Check first batch
for images, masks in train_loader:
  print(f"Batch of images shape: {images.shape}")
  print(f"Batch of masks shape: {masks.shape}")
  break
```

```
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models
import torch.nn.functional as F

# Check if GPU is available, otherwise use CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define model (U-Net model)
class UNet(nn.Module):
    def __init__(self, in_channels=3, out_channels=1):
        super(UNet, self).__init__()

        # Use ResNet34 as the encoder and access intermediate layers
        resnet = models.resnet34(weights='DEFAULT')

        # Encoder: Using layers up to the last convolutional block
        self.encoder1 = nn.Sequential(resnet.conv1, resnet.bn1, resnet.relu, resnet.maxpool)
        self.encoder2 = resnet.layer1  # Output ~64 channels
        self.encoder3 = resnet.layer2  # Output ~128 channels
        self.encoder4 = resnet.layer3  # Output ~256 channels
        self.encoder5 = resnet.layer4  # Output ~512 channels

        # Decoder: Build decoder with upsampling and concatenation (skip connections)
        # Decoder block takes input from previous decoder layer and corresponding encoder layer
        self.decoder4 = self._upconv(512 + 256, 256)  # 512 from encoder5, 256 from encoder4
        self.decoder3 = self._upconv(256 + 128, 128)  # 256 from decoder4, 128 from encoder3
        self.decoder2 = self._upconv(128 + 64, 64)  # 128 from decoder3, 64 from encoder2
        self.decoder1 = self._upconv(64 + 64, 64)  # 64 from decoder2, 64 from encoder1

        # Final output layer
        self.final_conv = nn.Conv2d(64, out_channels, kernel_size=1)

    def _upconv(self, in_channels, out_channels):
        # Helper function for a decoder block: Upsample and apply convolution
        return nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU()
        )

    def forward(self, x):
        # Encoder path
        enc1 = self.encoder1(x)
        enc2 = self.encoder2(enc1)
        enc3 = self.encoder3(enc2)
        enc4 = self.encoder4(enc3)
        enc5 = self.encoder5(enc4)  # Bottleneck layer

        # Decoder path with skip connections
        # Upsample dec5 and concatenate with enc4
        dec4_upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)(enc5)  # Upsample enc5
        dec4 = self.decoder4(torch.cat([dec4_upsample, enc4], dim=1))  # Concatenate bottleneck and encoder4 output

        # Upsample dec4 and concatenate with enc3
        dec3_upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)(dec4)  # Upsample decoder4 output
        dec3 = self.decoder3(torch.cat([dec3_upsample, enc3], dim=1))  # Concatenate decoder4 and encoder 3 output

        # Upsample dec3 and concatenate with enc2
        dec2_upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)(dec3)  # Upsample decoder3 output
        dec2 = self.decoder2(torch.cat([dec2_upsample, enc2], dim=1))  # Concatenate decoder3 and encoder2 output

        # Upsample dec2 and concatenate with enc1
        dec1_upsample = F.interpolate(dec2, size=enc1.shape[2:], mode='bilinear', align_corners=True)  # Upsample decoder2 output
        dec1 = self.decoder1(torch.cat([dec1_upsample, enc1], dim=1))  # Concatenate decoder and encoder1 output

        # Final output
        return self.final_conv(dec1)


# Initialize the model, optimizer, & loss function
model = UNet(in_channels=3, out_channels=1).to(device)  # Use the correct device (GPU or CPU)
optimizer = optim.Adam(model.parameters(), lr=1e-4)
criterion = nn.BCEWithLogitsLoss()

# Example of using the model (forward pass on the input data)
# Move your input data to the appropriate device (GPU or CPU)
# images, masks = images.to(device), masks.to(device)  # Example to move your data

# Note: You can load your dataset and pass the images to the model for training or inference
```

## Visualize prediction

Visualization of predictions on validation set

```
import matplotlib.pyplot as plt
import torch

# Function to denormalize the image
def denormalize(image, mean, std):
    for t, m, s in zip(image, mean, std):
        t.mul_(s).add_(m)  # Denormalize the image
    return image

# Check if GPU is available, otherwise use CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Set model to evaluation mode
model.eval()

# Move model to the appropriate device (GPU or CPU)
model = model.to(device)

# Get a batch of validation images and ground truth masks
for images, masks in val_loader:
    images, masks = images.to(device), masks.to(device)  # Move to GPU or CPU

    # Denormalize images to bring them back to [0, 1]
    images = denormalize(images, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])

    # Perform inference
    with torch.no_grad():
        outputs = model(images)

    # Convert predictions to binary masks
    predicted_masks = torch.sigmoid(outputs)
    predicted_masks = (predicted_masks > 0.5).float()

    # Display 1st image, predicted mask, & ground truth mask
    plt.figure(figsize=(15, 5))

    # Original Image
    plt.subplot(1, 3, 1)
    plt.imshow(images[0].cpu().numpy().transpose(1, 2, 0))  # Convert back to HWC format
    plt.title('Original Image')
    plt.axis('off')

    # Predicted Mask
    plt.subplot(1, 3, 2)
    plt.imshow(predicted_masks[0].cpu().numpy().squeeze(), cmap='gray')  # Remove channel dimension
    plt.title('Predicted Mask')
    plt.axis('off')

    # Ground Truth Mask
    plt.subplot(1, 3, 3)
    plt.imshow(masks[0].cpu().numpy().squeeze(), cmap='gray')  # Remove channel dimension
    plt.title('Ground Truth Mask')
    plt.axis('off')

    plt.show()
    break  # Visualize only the first batch
```
